```python
test_data = pd.read_csv('test_new.csv')
```

```python
X_t = np.array(test_data)
```

```python
le = LabelEncoder()
X_t[:,12] = le.fit_transform(X_t[:,12])
X_t[:,10] = le.fit_transform(X_t[:,10])
X_t[:,2] = le.fit_transform(X_t[:,2])
X_t[:,6] = le.fit_transform(X_t[:,6])
X_t[:,4] = le.fit_transform(X_t[:,4])

X_t[:,16] = le.fit_transform(X_t[:,16])
X_t[:,32] = le.fit_transform(X_t[:,32])
X_t[:,17] = le.fit_transform(X_t[:,17])

```

```python
import pickle
SVM_clf = pickle.load(open('Sigmoid.pkl', 'rb'))
```

```python
SVM_clf.predict(X_t)
SVM_pro = SVM_clf.predict_proba(X_t)
one_prob = SVM_pro[:,1]
one_prob
```

```python
import pickle
SVM_clf1 = pickle.load(open('Linear.pkl', 'rb'))
```

```python
a = SVM_clf1.predict(X_t)
p = SVM_clf1.predict_proba(X_t)
p[:,1]
```

```python
one_linear = p[:,1]
```

```python
mean_prob = []
```

```python
len(one_prob),len(one_linear)
```

```python
for i in range(len(one_prob)):
  mean_prob.append((one_prob[i]+one_linear[i])/2)
```

```python
mean_prob
```

```python
mean_dict = {}

for j in range(X_t.shape[0]):
  mean_dict[X_t[j][0]]={}
for j in range(X_t.shape[0]):
  mean_dict[X_t[j][0]]['Attrition']=mean_prob[j]
```

```python
regressor1 = pickle.load(open('RegressorNonScaled.pkl','rb'))
```

```python
y__1 = regressor1.predict(X_t)
```

```python
np.floor(y__1)
```

```python
RFR_clf = pickle.load(open('RegressionScaled.pkl','rb'))
```

```python
RFR_pred = RFR_clf.predict(X_t)
```

```python
RFR_pred
```

```python
y__1,RFR_pred
```

```python
mean_years = []
```

```python
for i in range(len(RFR_pred)):
  a = (np.floor(y__1[i])+np.floor(RFR_pred[i]))/2
  mean_years.append(np.floor(a))
```

```python
for j in range(X_t.shape[0]):
  mean_dict[X_t[j][0]]['Years'] = int(np.floor(mean_years[j]))
```

```python
final_test = []
```

```python
for i in range(X_t.shape[0]):
  list1 = []
  list1.append(X_t[i][0])
  list1.append(mean_dict[X_t[i][0]]['Attrition'])
  list1.append(mean_dict[X_t[i][0]]['Years'])
  final_test.append(list1)
```

```python
final = pd.DataFrame(final_test,columns = ('ID','Attrition','YearsAtCompany'))
final
```
